```python
import pandas as pd
from pandas import DataFrame as df
from IPython.display import display as dsply
import numpy as np

# This script can be used to check the various LSL tags recieved and handled by emotibit oscilloscope
# in tandem with the python Marker included in ofxLSL. The code most easily functions when the LSL stream data is printed
# to console by uncommenting the couts in:
#         void ofApp::update() found in ofApp.cpp of EmotiBit Oscilloscope
#         void ofApp::parseIncomingRequestData() of EmotiBit Oscilloscope
# Copy and paste the console stream into consoleDataLM.csv and remove everything but the LM lines
# This file has computer specific paths that should be changed to suit need

#Import Console Data
cols= ['Tag','LClab','LC','TSClab','TSC', 'TSlab', 'TS', 'Data']
data = pd.read_csv('c:/users/ctjen/desktop/consoleDataLM.csv',names=cols)
data = data.drop(columns=['Tag','LClab', 'TSClab','TSlab'])
#dsply(data)

#Import Python Outlet Stream
sdata = pd.read_csv('c:/users/ctjen/desktop/sentInfo.csv')
sdata = sdata.drop(columns=['I'])

#dsply(sdata)

# #Align and Format to Match the Console Data
#sdata = sdata.shift(-1)
for row in range(data['TS'].count()):
    if sdata.iloc[row,1] != data.iloc[row,3] :
        sdata = sdata.drop(row)
sdata=sdata.reset_index()
# sdata = sdata.drop([32,33])       #alignment fix
# sdata=sdata.reset_index()         #alignment fix
sdata = sdata.drop(columns=['index'])

#Append Python Data
data['tsPython']=sdata['Timestamp']

#Append Dif Cols
data['tsDif']= data['TS']-data['tsPython']

#Drop Tags
data = data.drop(columns=['Data'])
# data = data.drop(31)    #outlier due to alignment

#Deriv
data['deltaTSC']=data['TSC'].diff()
data['deltaLC']=data['LC'].diff()
data['deltaTSPython']=data['tsPython'].diff()
data['periodicity']=data['deltaTSC']-data['deltaTSPython']

#Display
pd.options.display.max_rows = None
dsply(data)
data.describe()
#data.to_csv('c:/users/ctjen/desktop/PythonAnalysis.csv')
#data.describe().to_csv('c:/users/ctjen/desktop/PythonAnalysisStat.csv')
#dsply(sdata)
```
